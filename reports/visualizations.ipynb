```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_final_elo = pd.read_csv("../data/processed/final_elo.csv")

# Sort by final Elo and take top 10
top_10 = df_final_elo.sort_values(by="final_elo", ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.bar(top_10["fighter"], top_10["final_elo"], color='skyblue')
plt.xlabel("Fighters")
plt.ylabel("Final Elo Rating")
plt.title("Top 10 UFC Fighters by Elo Rating")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df_final_elo["final_elo"], bins=30, color="purple", alpha=0.7, edgecolor="black")
plt.xlabel("Elo Rating")
plt.ylabel("Number of Fighters")
plt.title("Distribution of Fighter Elo Ratings")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
