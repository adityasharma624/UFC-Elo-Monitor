```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
k24 = pd.read_csv("../data/raw/fighter_elo_k24.csv")
k32 = pd.read_csv("../data/raw/fighter_elo_k32.csv")
k40 = pd.read_csv("../data/raw/fighter_elo_k40.csv")
```

```python
fighter = "Conor McGregor"
elo24 = k24[k24["fighter_1"] == fighter]
elo32 = k32[k32["fighter_1"] == fighter]
elo40 = k40[k40["fighter_1"] == fighter]
```

```python
elo24.loc[:, "date"] = pd.to_datetime(elo24["date"])
elo32.loc[:, "date"] = pd.to_datetime(elo32["date"])
elo40.loc[:, "date"] = pd.to_datetime(elo40["date"])
```

```python
elo24 = elo24.sort_values("date")
elo32 = elo32.sort_values("date")
elo40 = elo40.sort_values("date")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(elo24["date"], elo24["fighter_1_elo_after"], label="K=24", marker="o")
plt.plot(elo32["date"], elo32["fighter_1_elo_after"], label="K=32", marker="s")
plt.plot(elo40["date"], elo40["fighter_1_elo_after"], label="K=40", marker="^")

# Formatting
plt.xlabel("Date")
plt.ylabel("Elo Rating")
plt.title(f"Elo Rating Progression for {fighter}")
plt.legend()
plt.xticks(rotation=45)
plt.grid()

# Show plot
plt.show()
```
